# Джойним новости с финданными

```python
for my_date in tqdm(imoex_filtered_df['Date']):
    my_date = str(my_date)[:10]
    for topic_tag in unique_keys:
        news_records = filtered_lenta_df1.loc[(filtered_lenta_df1['date_dt']==my_date) &
                                              (filtered_lenta_df1['topic_tags']==topic_tag), 'title'].to_list()
        news_records = list(set(news_records))
        imoex_filtered_df.loc[imoex_filtered_df['Date']==my_date, topic_tag] = str(news_records)
```
