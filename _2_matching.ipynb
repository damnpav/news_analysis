```python
import pandas as pd
from tqdm import tqdm
import glob
```

```python
lenta_path = r'/Users/dp_user/Documents/универ/диплом/data/lenta/'
lenta_df = pd.read_csv(lenta_path + 'lenta-ru-news.csv')
```

```python
lenta_df['date_dt'] = pd.to_datetime(lenta_df['date'])
```

```python
lenta_df['week'] = lenta_df['date_dt'].dt.isocalendar().week
lenta_df['yr'] = lenta_df['date_dt'].dt.year
lenta_df['week_yr'] = lenta_df['week'].astype(str) + '_' + lenta_df['yr'].astype(str)
```

```python
# загрузим финансовые данные
```

```python
files_list = glob.glob('data/*')
```

```python
fin_data_dict = {}
for file in tqdm(files_list):
    current_df = pd.read_excel(file)
    fin_data_dict[file.replace('data/', '').replace('.xlsx', '')] = current_df
```
